```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# READ DATA
# oil
brent = pd.read_csv("./dataset/brent.csv")
wti = pd.read_csv("./dataset/wti.csv")

# oil related
oil_production  = pd.read_csv("./dataset/crudeoil_production1.csv") # oil production
oil_consumption     = pd.read_csv("./dataset/OECD_consumption_petroleum_1.csv") # oil consumption
oil_refined_consumption = pd.read_csv("./dataset/OECD_consumption_refinedpetroleum_1.csv") # refined oil consumption
oil_natural_gas  = pd.read_csv("./dataset/henryhubnaturalgas.csv") # natural gas
oil_product_tight  = pd.read_csv("./dataset/US-tight-oil-production.csv")
oil_stocks   = pd.read_csv("./dataset/total_stocks.csv")
oil_spr_stocks  = pd.read_csv("./dataset/SPR_stocks.csv")
oil_non_spr_stocks = pd.read_csv("./dataset/Non-SPR_stocks.csv")
oil_supply = pd.read_csv("./dataset/crudeoil_supplysummary.csv")

# price level
uscpi        = pd.read_csv("./dataset/USCPI.csv") # % data
uscpi_energy = pd.read_csv("./dataset/USCPI_energy.csv") # % data
usppi        = pd.read_csv("./dataset/USPPI.csv")
usppi_energy = pd.read_csv("./dataset/USPPI_energe.csv")
euppi        = pd.read_csv("./dataset/EUPPI.csv") # % data

# stocks
# date, open, high, low, close, adjclose, volume (n * 7)
snp          = pd.read_csv("./dataset/s&p.csv")
djia         = pd.read_csv("./dataset/djia.csv")
nasdaq       = pd.read_csv("./dataset/nasdaq.csv")
nyse         = pd.read_csv("./dataset/nyse.csv")

# ExR
euro_dollar  = pd.read_csv("./dataset/euro_dollar.csv")

# others
gold         = pd.read_csv("./dataset/gold.csv")
```

```python
# PREPROCESSING
# oil-related
oil_production = oil_production[["Month", "World"]]
oil_stocks = oil_stocks[["Date", "U.S. Ending Stocks of Crude Oil and Petroleum Products (Thousand Barrels)"]]
oil_non_spr_stocks = oil_non_spr_stocks[["Date", "U.S. Ending Stocks excluding SPR of Crude Oil and Petroleum Products (Thousand Barrels)"]]
oil_supply = oil_supply[["Date", "U.S. Field Production of Crude Oil (Thousand Barrels)"]]

# stocks : use close price
snp          = snp[["Date", "Close"]]
djia         = djia[["Date", "Close"]]
nasdaq       = nasdaq[["Date", "Close"]]
nyse         = nyse[["Date", "Close"]]

# ExRs
euro_dollar  = euro_dollar[["DATE", "DEXUSEU"]]
```

```python
dataset = {
    # oil
    "brent"        : brent,
    "wti"          : wti,
    
    "oil_production"  : oil_production, 
    "oil_consumption" : oil_consumption,
    "oil_refined_consumption" : oil_refined_consumption,
    "oil_natural_gas" : oil_natural_gas,
    "oil_product_tight" : oil_product_tight,
    "oil_stocks" : oil_stocks,
    "oil_spr_stocks" : oil_spr_stocks,
    "oil_non_spr_stocks" : oil_non_spr_stocks,
    "oil_supply" : oil_supply,

    # price level
    "uscpi"        : uscpi, 
    "uscpi_energy" : uscpi_energy,
    "usppi"        : usppi,
    "usppi_energy" : usppi_energy,
    "euppi"        : euppi,
    
    # stocks
    "snp"          : snp,
    "djia"         : djia,
    "nasdaq"       : nasdaq,
    "nyse"         : nyse,
    
    # ExRs
    "euro_dollar"  : euro_dollar,
    
    # others
    "gold"         : gold,
}
```

```python
# function to normalize dataset ; by minmax
def norm(x):
    _max = x.max()
    _min = x.min()
    _denom = _max - _min
    return (x - _min) / _denom
```

```python
for dataname, data in dataset.items():
    # data cleansing
    data.columns = ["DATE", "VALUE"]
    data['DATE'] = pd.to_datetime(data['DATE'], errors = 'coerce')
    data = data.set_index("DATE")
    data['VALUE'] = pd.to_numeric(data['VALUE'], errors = 'coerce')
    
    # normalize data
    data = norm(data)
    
    # set data range
    data = data["1980-01-01":"2020-03-01"]
    
    # Ploting
    isoil = (dataname == "brent") or (dataname == "wti")
    if(isoil):
        plt.plot(data, label = dataname, linewidth = 10)
    else:    
        plt.plot(data, label = dataname, linewidth = 3)
    
plt.legend(fontsize = 50, bbox_to_anchor=(1.04,1), loc="upper left")
plt.rc('font', size = 20)
plt.rcParams["figure.figsize"] = (50,25)
plt.show()
```
